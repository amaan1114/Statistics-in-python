```python
import statistics
import matplotlib.pyplot as plt
test_scores=[20,40,100,60,80,90,100,90,50,55,23,52,23,22,34,90]
#Calculating mean mode median
mean = statistics.mean(test_scores)
print(mean)
mode = statistics.mode(test_scores)
print(mode)
median = statistics.median(test_scores)
print(median)

#Visualizing it using histogram and axvline

plt.title("Graph of mean, mode, median of test scores")
data = [mean,mode,median]
plt.hist(test_scores, bins=10, edgecolor='black', color='lightblue')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean}')
plt.axvline(median, color='green', linestyle='--', label=f'Median: {median}')
plt.axvline(mode, color='blue', linestyle='--', label=f'Mode: {mode}')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
