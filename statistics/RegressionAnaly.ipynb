```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
#Years of Experience
X=np.array([1, 2, 3, 4,  6, 7, 8, 9, 10]).reshape(-1,1)
#Salary in thousands
y=np.array([30, 35, 40, 50, 58, 62, 70, 75, 85])

dataset = LinearRegression()
Regression = dataset.fit(X,y)

#Predicting salary of 5 years experience
pSalary  = dataset.predict([[5]])
print("The Predicted salary for the input is =",int(pSalary[0]),"k")



#Scatter plot graph
plt.title("experienc vs salary graph")
plt.scatter(X,y)
plt.ylabel("Salary")
plt.yticks(y,[f"{i}k" for i in y] )

plt.xlabel("Experience")
plt.plot(X,dataset.predict(X),color = "Red")

plt.show()
```
