```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Data
hours = np.array([1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0,
                  5.5, 6.0, 6.5, 7.0, 7.5, 8.0, 8.5, 9.0, 9.5, 10.0]).reshape(-1, 1)

marks = np.array([35, 40, 45, 50, 53, 58, 65, 68, 71,
                  75, 78, 82, 85, 88, 93, 95, 97, 99, 100]).reshape(-1,1)

#Scatter plot to visualize it
plt.title('Hours Studies vs Marks obtained')
plt.scatter(hours,marks)
plt.show()


```

```python
#calculating Corelation Coefficient.
cor = np.corrcoef(hours.ravel(),marks.ravel())[0,1]
print("Correlation coefficient: ",cor)
```

```python

#Fitting linear regression model and overlaying line on scatter plot
X_train,X_test,Y_train,Y_test = train_test_split(hours,marks,test_size=0.5)
model = LinearRegression()
model.fit(X_train,Y_train)


output = model.predict(X_test)

plt.plot(X_test,output)
plt.scatter(hours,marks)
plt.title('Hours Studies vs Marks obtained')
plt.show()

```
