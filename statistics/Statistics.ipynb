```python
import pandas as pd
import matplotlib.pyplot as plt
#Outliners 2000,4000
dataset = {'dataset':[20,30,50,50,60,2000,4000]}
dataset_framed = pd.DataFrame(dataset)

#Finding mean , mode ,median before outliners removed.
print("Mean before outliner removed: ",dataset_framed['dataset'].mean())
print("Median before outliner removed: ",dataset_framed['dataset'].median())
print("Mode before outliner removed: ",dataset_framed['dataset'].mode())

print()

#Finding mean,mode, median after outliners removed

dataset.update({'dataset':[20,30,50,50,60]})
dataset_framed2 = pd.DataFrame(dataset)

print("Mean After outliner removed: ",dataset_framed2['dataset'].mean())
print("Median After outliner removed: ",dataset_framed2['dataset'].median())
print("Mode After outliner removed: ",dataset_framed2['dataset'].mode())

# Comparison
Comparison = [dataset_framed['dataset'].mean(),dataset_framed2['dataset'].mean(),dataset_framed2['dataset'].mode(),dataset_framed2['dataset'].median(),dataset_framed['dataset'].mode(),dataset_framed['dataset'].median()]
plt.title("Comaprison graph of mean,mode,median after and before outliners removed\n")
plt.hist(Comparison,label=["Mean before outliner removed","Mean after outliner removed","Mode after outliner removed","Median after outliner removed","mode before outliner removed","Median before outliner removed"])
plt.legend()
plt.xlabel("mean,mode,median")
plt.ylabel("Density")
plt.show()

#Here we can see that mean is deviated due to outliners but mode and median are not.



```
