```python
import numpy as np
import matplotlib.pyplot as plt
#Dataset
test_scores=[20,40,100,60,80,90,100,90,50,55,23,52,23,22,34,90]
variance = np.var(test_scores)
sd = np.std(test_scores)
q1 = np.percentile(test_scores,25)
q3 = np.percentile(test_scores,75)
iqr = q3-q1


range = np.max(test_scores)-np.min(test_scores)
print("Range= ",range)
print("Iqr= ",iqr)
print("Variance= ",variance)
print("Standard Deviation= ",sd)
#Visualizing spread using boxplot

plt.title("Visualizing spread of test_data")
plt.ylabel("Scores")
plt.boxplot(test_scores)
plt.show()

#We can see that 100-90 is in 75 percentile 90 to 30 is in 50 percentile and 30 to 20 is in 25 percentile.
```
